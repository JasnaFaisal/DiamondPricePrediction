```python
#importing the necessary libraries
import pandas as pd
df=pd.read_csv('./data/gemstone.csv')
df.head()
```

```python
#dropping the unnecessary columns
df=df.drop('id',axis=1)
```

```python
df.head()
```

```python
#dividing into dependent and independent varaibles
#x=df.drop('price',axis=1)
#y=df[['price']]
```

```python
X = df.drop(labels=['price'],axis=1)
Y = df[['price']]
```

```python
X.head()
```

```python
Y.head()
```

```python
#segregating the numerical and categorical columns
numeric_col=X.select_dtypes(include='float').columns
categ_col=X.select_dtypes(exclude='float').columns
```

```python
numeric_col
```

```python
categ_col
```

```python
# Define the custom ranking for each ordinal variable
cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
#creating automated pipelines for feature engineering  and importing necessary libraries
from sklearn.impute import SimpleImputer         #handling missing values
from sklearn.preprocessing import OrdinalEncoder #handling ordinal encoding
from sklearn.preprocessing import StandardScaler #scaling
#pipelines
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

```

```python
#creating the numerical pipeline
num_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler())
    ]
)
```

```python
#categorical pipelines
categ_pipeline=Pipeline(
    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('ordinalencoder',OrdinalEncoder(categories=[cut_categories,color_categories,clarity_categories])),
        ('scaler',StandardScaler())
    ]
)
```

```python
categ_pipeline
```

```python
#column tranformer for concatenating the columns
preprocessor=ColumnTransformer([
('num_pipeline',num_pipeline,numeric_col),
('categ_pipeline',categ_pipeline,categ_col)
])

```

```python
preprocessor
```

```python
## Train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)
```

```python
X_train.head()
```

```python
X_test.shape
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

```python
## Model Training

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
## Train multiple models
## Model Ecaluation
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
    'Decisiontree':DecisionTreeRegressor()
}
```

```python


def evaluate_model(true, predicted):
    mse_value= mean_squared_error((true), (predicted))
    rmse_value =mean_squared_error((true), (predicted),squared=False)
    r2_square_value = r2_score((true), (predicted))
    mae_value=mean_absolute_error((true), (predicted))
    return mae_value, mse_value, rmse_value, r2_square_value
```

```python

#checking the scores for regressors

mae=[]
mse=[]
rmse=[]
r2score=[]
for name,model in models.items():
     model.fit(X_train,y_train)
     print(name+' trained')
     
     y_pred=model.predict(X_test)

     mae_value, mse_value, rmse_value, r2_square_value=evaluate_model(y_test,y_pred)
     mae.append(mae_value)
     mse.append(mse_value)
     rmse.append(rmse_value)
     r2score.append(r2_square_value)
```

```python
Models = ['Linear Regressor','Lasso Regressor','Ridge Regressor','ElasticNet Regressor','DecisionTree Regressor']
model_performance = pd.DataFrame({
    'Model':Models,
    'R2_Score':r2score,
    'Mean_squared_error':mse,
    'Root_Mean_Squared_Error':rmse,
    'Mean_Absolute_error':mae
    })
model_performance
```
